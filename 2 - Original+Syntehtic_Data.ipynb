# Analysis of the synthetic dataset - Creation of the merged dataset

```python
import pandas as pd
import numpy as np
```

```python
synthetic_data = pd.read_csv("Breast Cancer Dataset Augmentation_synthetic.csv", sep=';')
original_data = pd.read_csv("data.csv")
```

```python
synthetic_data.shape
```

```python
original_data = original_data.drop(['Unnamed: 32'], axis=1)
original_data.shape
```

```python
merged_data = pd.concat([original_data, synthetic_data], axis=0)
```

```python
merged_data.shape
```

```python
description = merged_data.drop(['id'], axis=1)
description = description.describe().transpose()
description
```
